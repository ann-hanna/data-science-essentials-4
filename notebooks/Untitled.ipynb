```python
hospitalizations = pd.read_csv('../Data/Medicare_Disparities_by_Population/hospitalizations.csv')
```

```python
hospitalizations=hospitalizations[["county","state","urban","primary_denominator","analysis_value"]]
hospitalizations = hospitalizations.loc[hospitalizations['state'] == 'TENNESSEE']
hospitalizations = hospitalizations.rename(columns={'analysis_value': 'hospitalizations_per_1000'})
hospitalizations.head()
```

```python
emergency_department = pd.read_csv('../Data/Medicare_Disparities_by_Population/emergency_department.csv')
emergency_department.head()
```

```python
emergency_department=emergency_department[["county","state","urban","analysis_value"]]
emergency_department = emergency_department.loc[emergency_department['state'] == 'TENNESSEE']
emergency_department = emergency_department.rename(columns={'analysis_value': 'er_visits_per_1000'})
emergency_department.head()
```

```python
annual_wellness = pd.read_csv('../Data/Medicare_Disparities_by_Population/annual_wellness.csv')
annual_wellness.head()
```

```python
annual_wellness=annual_wellness[["county","state","urban","analysis_value"]]
annual_wellness = annual_wellness.loc[annual_wellness['state'] == 'TENNESSEE']
annual_wellness = annual_wellness.rename(columns={'analysis_value': 'wellness_visits_percentage'})
annual_wellness.head()
```

```python
hospitalizations_er_visits= pd.merge(left = hospitalizations,
         right = emergency_department,
         on='county',
        how='left'
        )
hospitalizations_er_visits.head()
```

```python
hospitalizations_er_visits_vs_annual_wellness= pd.merge(left = hospitalizations_er_visits,
                                                         right = annual_wellness,
                                                        on='county'
                                                       )
hospitalizations_er_visits_vs_annual_wellness
```

```python
hospitalizations_er_visits_vs_annual_wellness[["county","urban","hospitalizations_per_1000","er_visits_per_1000","wellness_visits_percentage"]]
```
